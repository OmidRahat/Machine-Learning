```python
df=pd.read_csv("/content/drive/MyDrive/Machin Learning/diabetes.csv")
df.head()
```

```python
df["Outcome"].value_counts()
```

```python
x=df.drop("Outcome",axis=1)
x
y=df["Outcome"]
y
```

```python

x=np.array(x)
x
y=np.array(y)
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier(max_depth=18,min_samples_split=2,min_samples_leaf=1)
dt.fit(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier(n_estimators=200,max_depth=18,min_samples_split=2,min_samples_leaf=1)
rf.fit(x_train,y_train)
```

```python
#DecisionTree
dt_test_predict=dt.predict(x_test)
dt_train_predict=dt.predict(x_train)
#Random Forest
rf_test_predict=rf.predict(x_test)
rf_train_predict=rf.predict(x_train)
```

```python
from sklearn.metrics import precision_score,recall_score,accuracy_score,confusion_matrix
```

```python
ps_dt=precision_score(y_test,dt_test_predict)
ps_rf=precision_score(y_test,rf_test_predict)
ps_dt,ps_rf
```

```python
ps_dt=precision_score(y_train,dt_train_predict)
ps_rf=precision_score(y_train,rf_train_predict)
ps_dt,ps_rf
```

```python
dt_test_acc=accuracy_score(y_test,dt_test_predict)
dt_train_acc=accuracy_score(y_train,dt_train_predict)

rf_test_acc=accuracy_score(y_test,rf_test_predict)
rf_train_acc=accuracy_score(y_train,rf_train_predict)

dt_test_acc,dt_train_acc,rf_test_acc,rf_train_acc
```

```python
dt_rc_test=recall_score(y_test,dt_test_predict)
dt_rc_train=recall_score(y_train,dt_train_predict)

rf_rc_test=recall_score(y_test,rf_test_predict)
rf_rc_train=recall_score(y_train,rf_train_predict)

dt_rc_test,dt_rc_train,rf_rc_test,rf_rc_train
```

```python
dt_cm_test=confusion_matrix(y_test,dt_test_predict)
dt_cm_train=confusion_matrix(y_train,dt_train_predict)

rf_cm_test=confusion_matrix(y_test,rf_test_predict)
rf_cm_train=confusion_matrix(y_train,rf_train_predict)

dt_cm_test,dt_cm_train,rf_cm_test,rf_cm_train
```
