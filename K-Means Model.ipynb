Create a sample data for our model

```python
from sklearn.datasets import make_blobs
x,y=make_blobs(n_samples=300,n_features=2,centers=3)
x.shape
```

to see our sample

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,1])
plt.show()
```

creat Kmean model

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=3)
model.fit(x)
```

```python
center=model.cluster_centers_
lable=model.labels_
```

See our model clustering

```python
plt.scatter(x[:,0],x[:,1],c=lable)
plt.scatter(center[:,0],center[:,1],c='r',marker='*')
plt.show()
```
