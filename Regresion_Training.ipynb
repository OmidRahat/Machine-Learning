```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([1,2,3,4])
y=np.array([1,3,5,7])
plt.plot(x,y,"o",color="red")
plt.show()
```

```python
x
```

```python
model=LinearRegression()
model.fit(x.reshape(-1,1),y.reshape(-1,1))
plt.plot(x,model.predict(x.reshape(-1,1)),marker="o",color="red",linestyle="dashed")
plt.show()
```

```python
model=LinearRegression()
model.fit(np.array(x).reshape(-1,1),np.array(y).reshape(-1,1))
plt.plot(x,y,"o",color="red")
plt.plot(x,model.predict(np.array(x).reshape(-1,1)),color="blue")
plt.show()
```

```python
#ba tavajoh be x ha bebinim agar vorodi 5 bashe model chi pish bini mikone
result=model.predict([[5],[6],[7]])
result
```
